```python
import numpy as np
```

```python
def slc_max(a,k):
    max=abs(a[k-1])
    for i in range(k-1,np.shape(a)[0]):
        if abs(a[i])>=max:
            j=i
    return j
```

```python
def L_func(k,A):
    n=np.shape(A)[0]
    L=np.zeros(n*n).reshape(n,n)
    for i in range(n):
        L[i][i]=1
        if i>k-1:
            L[i][k-1]=-A[i][k-1]/A[k-1][k-1]
    return L
```

```python
#对矩阵进行行交换所用的函数（交换第k行和第j+1行）
def I_line_exchange(k,j,A):
    #j为第k列最大列元所在行
    n=np.shape(A)[0]
    I=np.zeros(n*n).reshape(n,n)
    for i in range (n):
        I[i][i]=1
    I[k-1][k-1]=0
    I[j][j]=0
    I[j][k-1]=1
    I[k-1][j]=1
    return I
```

```python
def col_main(A,b):
    for i in range(np.shape(A)[0]):
        I=I_line_exchange(i+1,slc_max(A[:,i],i+1),A)
        A=np.dot(I,A)
        b=np.dot(I,b)
        L=L_func(i+1,A)
        A=np.dot(L,A)
        b=np.dot(L,b)
    n=np.shape(A)[0]
    x=np.zeros(n).reshape(n,1)
    x[-1]=b[-1]/A[-1][-1]
    for i in range(n-2,-1,-1):
        sum=0
        for j in range(i+1,n):
            sum+=A[i][j]*x[j]
        x[i]=(b[i]-sum)/A[i][i]
    return A,x
```

```python
#代入数据
A=np.ones(400*400).reshape(400,400)
b=np.zeros(400*1).reshape(400,1)
for i in range(400):
    A[i][i]=399+i+1
    b[i]=(399+i+1)*(i+1)+80200-(i+1)
col_main(A,b)
```
